```python
import torch
import cv2
```

```python
model = torch.hub.load('WongKinYiu/yolov7', 'custom', '10.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture('videos/car_sample.mp4')
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('car.mp4', fourcc, 30, (int(cap.get(3)), int(cap.get(4))))
```

```python
import pandas as pd
```

```python
column_names = ["Frame", "x1", "y1", "x2", "y2", "Confidence", "Vehicle"]

df = pd.DataFrame(columns = column_names)
```

```python
i = 0
while True:
    
    ret, frame = cap.read()
    
    if frame is not None:
    
        results = model(frame)

        for result in results.xyxy[0]:
            x1, y1, x2, y2, conf, cls = result.tolist()
            
            if cls == 0:
                name = 'Bike'
            elif cls == 1:
                name = 'Bus'
            else:
                name = 'Car'
            
            df = df.append({'Frame':i, 'x1':x1, 'y1':y1, 'x2':x2, 'y2':y2, 'Confidence':conf, 'Vehicle':name}, ignore_index=True)
            
            df = df.round(1)
            
            if conf > 0.7:
                if cls == 0:
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                    cv2.putText(frame, f'{model.names[int(cls)]} {conf:.2f}', (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 0), 2)
                if cls == 1:
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
                    cv2.putText(frame, f'{model.names[int(cls)]} {conf:.2f}', (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 0), 2)
                if cls == 2:
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                    cv2.putText(frame, f'{model.names[int(cls)]} {conf:.2f}', (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 0), 2)
        
        cv2.imshow('frame', frame)
        out.write(frame)
        
        i = i +1
        if i % 100 == 0:
            print('frame' ,i)

        if cv2.waitKey(1) == ord('q'):
            break
        
    else:
        print('Detecting Ended')
        df.to_csv('Car.csv')
        cap.release()
        out.release()
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
